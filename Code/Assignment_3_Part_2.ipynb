```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('A:\\Python\\titanic.csv')
```

__Below we are going to determine the survival rate of passengers based on their class.__

```python
survived_by_class=df.groupby('Pclass')['Survived'].mean()
sns.barplot(x=survived_by_class.index,y=survived_by_class.values)
plt.xlabel('Class of Passengers')
plt.ylabel('Survival rate of passengers')
plt.title('Survival rate based on passenger class')
```

```python
sns.countplot(data=df, x='Pclass', hue='Survived')
plt.xlabel('Class of Passengers')
plt.title('Survival and death rate based on passenger class')
```

```python
df['Pclass'].corr(df['Survived'])
```

__In below hypothesis we are going to determine if the survival rate is associated to the gender__

```python
survival_gender=df.groupby('Sex')['Survived'].mean()
sns.barplot(x=survival_gender.index,y=survival_gender.values)
plt.xlabel('Gender of the passenger')
plt.ylabel('Survival rate of the passenger')
plt.title('Survival rate based on Gender of passenger')
```

```python
sns.countplot(data=df, x='Sex', hue='Survived')
```

__Below we are going to determine if the survival rate is associated to the age__

```python
age_grp=[0,18,60,100]
age_label=['Children','Adult','Senior']
df['Age_group']=pd.cut(df['Age'],bins=age_grp,labels=age_label)
survival_age=df.groupby('Age_group')['Survived'].mean()
sns.barplot(x=survival_age.index,y=survival_age.values)
plt.xlabel("Age group of the passenger")
plt.ylabel("Survival rate of passenger")
plt.title("Survival rate of passenger based on the Age group")
```

```python
sns.countplot(data=df, x='Age_group', hue='Survived')
```
